```python
import json
import re
import nltk
import itertools
import matplotlib.pyplot as plt
import HTMLParser
from collections import Counter
import numpy as np
%matplotlib inline
```

```python
# Load all data from Dice, one JSON object per line
dice_records = open("../data/dice_processed/jobs.jsonlines").readlines()

def load_json_save(x):
    try:
        return json.loads(x)
    except ValueError, e:
        pass
    return None;

# Filter out records that couldn't be parsed
dice_records = map(lambda r: load_json_save(r), dice_records)
dice_records = filter(lambda r: r, dice_records)
```

```python
# Preprocessing: 
# - Downcase the text
# - Replace \xa0 (non-breaking space) characters
# - Replace \u2019 (single quotation mark) characters
# - Replace \u2019 (double quotation mark) characters
# - Replace \u2022 (bullet) characters
# - Tokenize using NLTK (for cleanup)
h = HTMLParser.HTMLParser()
for record in dice_records:
    record["description"] = record["description"].lower()
    record["description"] = h.unescape(record["description"])
    record["description"] = record["description"].replace(u'\xa0', u' ')
    record["description"] = record["description"].replace(u'\u2022', '')
    record["description"] = record["description"].replace(u'\u2019', u"'")
    record["description"] = record["description"].replace(u'\u2013', u"\"")
    
#     sentences = nltk.sent_tokenize(record["description"])
#     sentences = [nltk.word_tokenize(sent) for sent in sentences]
#     sentences = [" ".join(sent) for sent in sentences]
#     record["description"] = " ".join(sentences)
    
```

```python
def is_remote_job(record):
    if "remote" in record["title"].lower():
        return True
    elif "telecommute" in record["title"].lower():
        return True
    return False

remote_jobs = filter(lambda x: is_remote_job(x[1]), enumerate(dice_records))
remote_job_indices = [x[0] for x in remote_jobs]
remote_jobs = [x[1] for x in remote_jobs]
print "Found %d remote/telecommute jobs" %(len(remote_jobs))
```

```python
# Data Exploration: Find the job locations
loc_counter_all = Counter()
loc_counter_all.update([r["location"] for r in dice_records])
print "All:"
print loc_counter.most_common(20);

# Data Exploration: Find the job locations
loc_counter_remote = Counter()
loc_counter_remote.update([r["location"] for r in remote_jobs])
print "Remote:"
print loc_counter.most_common(20);
```

```python
# Plot the locations for all jobs
fig, ax = plt.subplots(figsize=(12,5))
N = 20
data = [x[1] for x in loc_counter_all.most_common(N)]
rects = ax.bar(np.arange(N), data, 0.5)
ax.set_title("Locations (%f percent coverage)"%(100.0*sum(data)/float(sum(loc_counter_all.values()))))
ax.set_xticks(np.arange(N))
ax.set_xticklabels([x[0] for x in loc_counter.most_common(N)], rotation=90)
plt.show()
```

```python
# Looks for common technologies
# I stole most of this from http://codingvc.com/which-technologies-do-startups-use-an-exploration-of-angellist-data
programming_languages = ["javascript", "ruby", "php", "node.js", "java", "scala", "python", "c++", "perl", "clojure", ["c#","c #"]]
storage = ["mysql", ["postgresql", "postgres"], "mongodb", "redis", "hdfs", "sql server", 
           "memcached", "cassandra", "dynamodb", "hbase", "amazon s3"]
frontend = ["html", "css", "jquery", "angular.js", "backbone.js", "rails", "django", "express.js", "ember.js", "d3.js"]
hosting = [["aws", "amazon web services", "ec2"], "heroku", "linode", "digital ocean", "azure"]
def term_counts(terms, data):
    counter = Counter()
    for term in terms:
        if isinstance(term, str):
            counter[term] = sum(1 for x in data if " %s "%term in x["description"])
        elif isinstance(term, list):
            for variation in term:
                counter[term[0]] += sum(1 for x in data if " %s "%variation in x["description"])
    return counter 

def plot_term_counts(counter, title=""):
    fig, ax = plt.subplots(figsize=(10,5))
    N = len(counter)
    sorted_items = sorted(counter.items(), key=lambda x: x[1], reverse=True)
    x_labels, y_values = zip(*sorted_items)
    x_values = np.arange(N)
    rects = ax.bar(x_values, y_values, 0.5)
    ax.set_title(title)
    ax.set_xticks(x_values)
    ax.set_xticklabels(x_labels, rotation=90)
    plt.show()

plot_term_counts(term_counts(programming_languages, dice_records), "Programming languages (all jobs)")
plot_term_counts(term_counts(programming_languages, remote_jobs), "Programming languages (remote jobs)")
plot_term_counts(term_counts(storage, dice_records), "Storage technologies (all jobs)")
plot_term_counts(term_counts(storage, remote_jobs), "Storage technologies (remote jobs)")
plot_term_counts(term_counts(frontend, dice_records), "Frontend technologies (all jobs)")
plot_term_counts(term_counts(frontend, remote_jobs), "Frontend technologies (remote jobs)")
plot_term_counts(term_counts(hosting, dice_records), "Hosting providers (all jobs)")
plot_term_counts(term_counts(hosting, remote_jobs), "Hosting providers (remote jobs)")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer

# Use a custom tokenizer from NLTK to get cleaer results
def tokenize(doc):
    sentences = nltk.sent_tokenize(doc)
    sentences = [nltk.word_tokenize(sent) for sent in sentences]
    return itertools.chain(*sentences)

# Count the terms in each description
print "Generating term counts for descriptions..."
vectorizer_desc = CountVectorizer(min_df=2, stop_words="english", tokenizer=tokenize, ngram_range=(2,2))
counts_all_desc = vectorizer_desc.fit_transform(map(lambda x: x["description"], dice_records))
counts_remote_desc = [counts_all_desc[i] for i in remote_job_indices]
# Generate TFIDF vectors
print "Generating tfidf vectors for descriptions..."
transformer = TfidfTransformer(norm="l1")
tfidf_all_desc = transformer.fit_transform(counts_all_desc)
tfidf_remote_desc = [tfidf_all_desc[i] for i in remote_job_indices]

# Count the terms in each title
print "Generating term counts for titles..."
vectorizer_title = CountVectorizer(min_df=2, stop_words="english")
counts_all_title = vectorizer_title.fit_transform(map(lambda x: x["title"], dice_records))
counts_remote_title = [counts_all_title[i] for i in remote_job_indices]
# Generate TFIDF vectors
print "Generating tfidf vectors for titles..."
transformer = TfidfTransformer(norm="l1")
tfidf_all_title = transformer.fit_transform(counts_all_title)
tfidf_remote_title = [tfidf_all_title[i] for i in remote_job_indices]

print "Done."
```

```python
def get_top_terms(vectorizer, data, N=20):
    features = vectorizer.get_feature_names()
    top_features = [sorted(zip(record.indices, record.data), key=lambda x: x[1], reverse=True)[:25] for record in data] 
    top_terms = map(lambda x: [features[i[0]] for i in x], top_features)
    counter = Counter()
    counter.update(itertools.chain(*top_terms))
    return counter.most_common(N)

print get_top_terms(vectorizer_desc, tfidf_all_desc, 50)

# print get_top_terms(vectorizer_title, tfidf_all_title)
# print get_top_terms(vectorizer_title, [tfidf_all_title[i] for i in remote_job_indices])
# print get_top_terms(vectorizer_desc, [tfidf_all_desc[i] for i in remote_job_indices], 50)
```
